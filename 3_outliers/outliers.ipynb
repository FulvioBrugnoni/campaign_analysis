```python
import numpy as np

np.random.seed(42)

data1 = np.random.normal(loc=300, scale=500, size=1000)
data2 = np.random.normal(loc=300, scale=500, size=1000)

print("Primi 10 valori di data1:", data1[:10])
print("Primi 10 valori di data2:", data2[:10])
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def plot_histogram_with_normal_distribution(data1, data2):

    plt.figure(figsize=(10, 6))

    sns.distplot(data1, color='blue', label='Data 1', kde=True, fit=None)
    mean_data1 = np.mean(data1)
    median_data1 = np.median(data1)
    plt.axvline(x=mean_data1, color='blue', linestyle='--', label=f'Data 1 Mean: {mean_data1:.2f}')
    plt.axvline(x=median_data1, color='blue', linestyle='-', label=f'Data 1 Median: {median_data1:.2f}')

    sns.distplot(data2, color='lime', label='Data 2', kde=True, fit=None)
    mean_data2 = np.mean(data2)
    median_data2 = np.median(data2)
    plt.axvline(x=mean_data2, color='lime', linestyle='--', label=f'Data 2 Mean: {mean_data2:.2f}')
    plt.axvline(x=median_data2, color='lime', linestyle='-', label=f'Data 2 Median: {median_data2:.2f}')

    plt.xlabel('Values')
    plt.ylabel('Density')
    plt.title('Overlaid Histogram with Normal Distribution')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
plot_histogram_with_normal_distribution(data1, data2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_histogram_with_boxplot(data1, data2):
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    sns.distplot(data1, color='blue', label='Data 1', kde=True, fit=None)
    plt.xlabel('Values')
    plt.ylabel('Density')
    plt.title('Data 1 Distribution')
    plt.legend()
    plt.grid(True)

    plt.subplot(1, 2, 1)
    sns.boxplot(data=data1, color='blue', width=0.2)

    plt.subplot(1, 2, 2)
    sns.distplot(data2, color='lime', label='Data 2', kde=True, fit=None)
    plt.xlabel('Values')
    plt.ylabel('Density')
    plt.title('Data 2 Distribution')
    plt.legend()
    plt.grid(True)

    plt.subplot(1, 2, 2)
    sns.boxplot(data=data2, color='lime', width=0.2)

    plt.tight_layout()
    
    plt.show()
```

```python
plot_histogram_with_boxplot(data1, data2)
```

```python
data1 = np.random.normal(loc=300, scale=500, size=1000)

data2 = np.concatenate((np.random.normal(loc=300, scale=500, size=990), [-1000000], 
                        [-1000000]
                       , [-1000000]
                       , [-1000000]
                       , [-1000000]
                       , [-1000000]
                       , [-1000000]
                       , [-1000000]
                       , [-1000000]
                       , [-1000000]))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.figure(figsize=(10, 6))

sns.distplot(data1, color='blue', label='Data 1', kde=True, fit=None)
mean_data1 = np.mean(data1)
median_data1 = np.median(data1)
plt.axvline(x=mean_data1, color='blue', linestyle='--', label=f'Data 1 Mean: {mean_data1:.2f}')
plt.axvline(x=median_data1, color='blue', linestyle='-', label=f'Data 1 Median: {median_data1:.2f}')

sns.distplot(data2, color='lime', label='Data 2', kde=True, fit=None)
mean_data2 = np.mean(data2)
median_data2 = np.median(data2)
plt.axvline(x=mean_data2, color='lime', linestyle='--', label=f'Data 2 Mean: {mean_data2:.2f}')
plt.axvline(x=median_data2, color='lime', linestyle='-', label=f'Data 2 Median: {median_data2:.2f}')

plt.xlim(-1000, 1000)

plt.xlabel('Values')
plt.ylabel('Density')
plt.title('Overlaid Histogram with Normal Distribution')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.stats import shapiro

stat_data1, p_data1 = shapiro(data1)
print("Shapiro-Wilk Test for Normality - Data 1:")
print("Test Statistic =", round(stat_data1, 2))
print("p-value =", round(p_data1, 2))

stat_data2, p_data2 = shapiro(data2)
print("\nShapiro-Wilk Test for Normality - Data 2:")
print("Test Statistic =", round(stat_data2, 2))
print("p-value =", round(p_data2, 2))
```

```python
from scipy.stats import ttest_ind

statistic, p_value = ttest_ind(data1, data2)

print("Independent t-test results:")
print("Test Statistic =", statistic.round(2))
print("p-value =", p_value.round(2))
```

```python
from scipy.stats import mannwhitneyu

statistic, p_value = mannwhitneyu(data1, data2)

print("Mann-Whitney U test results:")
print("Test Statistic =", statistic.round(2))
print("p-value =", p_value.round(2))
```
